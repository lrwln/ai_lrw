````python
# 参数默认值 ，es6 借鉴了，函数灵活
# llm 很强大， 封装的特别好， 就是几个接口
# completion 接口 aigc  prompt
# chat 聊天接口 messages 多轮会话
# chat 分角色  user, assistant

def get_completion(prompt, model='deepseek-chat'):
    response = client.chat.completions.create(
        model = model,
        messages = [
            { "role": "user", "content": prompt }
        ],
        # 自由度参数 0 严谨 2 嘻哈
        temperature=0.5,
        # 问题+回复的token数量
        max_tokens=1024,
    )
    # print(response)
    return response.choices[0].message.content

print(get_completion("写一首夏天游玩的七言诗"))
    
````

````python
text = f"""
你应当通过尽可能清晰、具体的指令，来明确你希望模型完成的任务。
这能引导模型产出符合预期的结果，同时降低回复内容偏离主题或出现错误的概率。
不要把编写清晰的提示词和精简提示词混为一谈。
很多时候，篇幅更长的提示词能为模型提供更完整的说明与背景信息，进而让输出内容更加详实、贴合需求。
"""
prompt = f"""
将三个反引号之间的文本总结为一句话
```{text}```
"""
prompt
print(get_completion(prompt))
````
